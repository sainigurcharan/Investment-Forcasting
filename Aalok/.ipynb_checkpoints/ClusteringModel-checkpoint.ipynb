```python
sector_averages2 =sector_averages.drop('Date', axis=1)
sector_averages2
```

```python
for col in sector_averages2.columns: 
    print(col) 
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Sector Dendrograms")
dend = shc.dendrogram(shc.linkage(sector_averages2, method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5
                                  
                                  , affinity='euclidean', linkage='complete')
cluster_h = cluster.fit_predict(sector_averages2)
cluster_h
```

```python
cluster_Hierarchical = pd.DataFrame(cluster_h)
cluster_Hierarchical
```

```python
cluster_data=sector_averages
cluster_data['cluster'] = cluster_Hierarchical
cluster_data
```

```python
cluster_values = sector_data.merge(cluster_data,on='Date', how ='left')
cluster_values
```
