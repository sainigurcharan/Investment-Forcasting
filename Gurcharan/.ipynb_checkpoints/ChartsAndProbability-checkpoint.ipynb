```python
import pandas as pd

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image  
import PIL
```

```python
chart_data = pd.read_csv(f"AllDataProbability.csv")
```

```python
chart_data
```

```python

lr_up = sns.relplot(x="LR_Up_Prob", y="Target_FR12Mts", 
            sizes=(400, 4000), alpha=1, palette="muted",
            height=6, data=chart_data)
plt.title("Logistics Regression / Forward 12M Returns > 8%")
plt.axhline(0.08, color="green", linestyle="--")
plt.axhline(0.00, color="red", linestyle="--")
plt.axvline(0.5, color="red", linestyle="--")
plt.axvline(0.75, color="green", linestyle="--")

# plt.show(lr_up)
plt.savefig('lr_up.jpg')
plt.savefig('lr_up.jpg', bbox_inches='tight')
```

```python

lr_down = sns.relplot(x="LR_Down_Prob", y="Target_FR12Mts", 
            sizes=(400, 4000), alpha=1, palette="muted",
            height=6, data=chart_data)
plt.title("Logistics Regression / Forward 12M Returns < 0%")

plt.axhline(0.00, color="red", linestyle="--")
plt.axvline(0.5, color="red", linestyle="--")
plt.axvline(0.75, color="green", linestyle="--")

# plt.show(lr_down)
plt.savefig('lr_down.jpg')
plt.savefig('lr_down.jpg', bbox_inches='tight')
```

```python
from matplotlib.colors import ListedColormap
flatui = ["#e74c3c", "#34495e","#95a5a6","#2ecc71","#9b59b6", "#3498db", ]          
my_cmap = ListedColormap(sns.color_palette(flatui).as_hex())
y = chart_data['RF1_Up_Prob']
x = chart_data['RF2_Up_Prob']
z = chart_data['Target_FR12Mts']
colors = chart_data['Target_FR12Mts']
# Plot
plt.figure(figsize=(12, 5))
sc = plt.scatter(x, y, c=colors, cmap=my_cmap)
                 
plt.title('RFModels vs. 12MonthReturn')
plt.xlabel('RF2_Up_Prob')
plt.ylabel('RF1_Up_Prob')
plt.axhline(0.26, color="blue", linestyle="--")
plt.axvline(0.43, color="blue", linestyle="--")

plt.axhline(0.5, color="red", linestyle="--")
plt.axvline(0.5, color="red", linestyle="--")
plt.colorbar(sc)
# plt.show()

plt.savefig('Rf_Both_Up.jpg')
plt.savefig('Rf_Both_Up.jpg', bbox_inches='tight')
```
